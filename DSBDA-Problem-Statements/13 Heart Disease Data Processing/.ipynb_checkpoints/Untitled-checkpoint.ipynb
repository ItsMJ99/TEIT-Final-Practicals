# Data cleaning(Remove NA, ?, Negative values etc.)

```python
#checking the data type of different values

for col in data.columns:
    print(col,data[col].unique())
    print('\n')
```

```python
data.head(1)
```

```python
# data in above can be seen in form string, int and float mixture so we need to make it of similar datatype
#error='coerce' means to ignore any error warning related to conversion
#The errors='coerce' parameter ensures that any values that cannot be converted to numeric format will be replaced with NaN (missing values)

data['trestbps']=pd.to_numeric(data['trestbps'],errors='coerce')
data['chol']=pd.to_numeric(data['chol'],errors='coerce')
data['fbs']=pd.to_numeric(data['fbs'],errors='coerce')
data['restecg']=pd.to_numeric(data['restecg'],errors='coerce')
data['thalach']=pd.to_numeric(data['thalach'],errors='coerce')
data['exang']=pd.to_numeric(data['exang'],errors='coerce')
data['oldpeak']=pd.to_numeric(data['oldpeak'],errors='coerce')
data['slope']=pd.to_numeric(data['slope'],errors='coerce')
data['ca']=pd.to_numeric(data['ca'],errors='coerce')
data['thal']=pd.to_numeric(data['thal'],errors='coerce')
```

```python
for col in data.columns:
    print(col,data[col].unique())
```

```python
(data == '?').sum()
```

```python
data.isna().sum()
```

```python
#as there are multiple varity of values in some columns so we will fill the missing values with mean
#as there are few variety of values in some columns so we will fill the missing values with mode (max occuring value)

cols_fill_mean = ['trestbps','chol','thalach','oldpeak']
mean_values = data[cols_fill_mean].mean()
data = data.fillna(mean_values)
data.isna().sum()
```

```python
data.shape
```

```python
cols_fill_mode = ['fbs','restecg','exang','slope','ca','thal']

for col in cols_fill_mode:
    data=data.fillna(data[col].mode()[0])

data.isna().sum()
```

# Error correcting(Outlier detection and removal)

```python
def outlier_detect_remove(data, threshold=3):
    zscore = np.abs((data - data.mean()) / data.std())
    out = zscore > threshold
    data_no_outliers = data[~out.any(axis=1)]
    data_outliers = data[out.any(axis=1)]
    return data_no_outliers, data_outliers

filtered_data,outliers = outlier_detect_remove(data)
```

# Data transformation
#### One-Hot Encoding:
##### One-hot encoding is used to convert categorical variables into a binary matrix, where each category becomes a separate binary feature.
##### Each category is represented by a binary column, where a value of 1 indicates the presence of the category and 0 indicates absence.
##### This technique is useful when there is no ordinal relationship between categories.
#### Label Encoding:
##### Label encoding is used to convert categorical variables into numerical labels.
##### Each category is assigned a unique integer value.
##### This technique is suitable when there is an ordinal relationship between categories.

# Build Data model using regression and kNN methods and compare accuracy of heart disease prediction.

```python
x = data[['age','sex','cp','oldpeak','ca','thal',]]
y = data['target']

x_train,x_test,y_train,y_test =  train_test_split(x,y,train_size=0.7, random_state=69)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
log_reg_model = LogisticRegression()
log_reg_model.fit(x_train,y_train)
```

```python
log_reg_pred = log_reg_model.predict(x_test)
```

```python
log_reg_accuracy = accuracy_score(y_test,log_reg_pred)
log_reg_accuracy
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(x_train,y_train)
```

```python
knn_pred = knn_model.predict(x_test)
```

```python
knn_accuracy=accuracy_score(y_test,knn_pred)
knn_accuracy
```

# Comparison

```python
print('Accuracy of Logistic Regression : {}'.format(log_reg_accuracy))
print('Accuracy of KNeighborsClassifier : {}'.format(knn_accuracy))
```
