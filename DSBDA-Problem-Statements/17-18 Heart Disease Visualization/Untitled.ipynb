```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cleavland_data = pd.read_csv('./Cleavland.csv',header=None)
cleavland_data
```

```python
hung_data = pd.read_csv('./hung.csv',header=None)
hung_data
```

```python
switzerland_data = pd.read_csv('./Switzerland.csv',header=None)
switzerland_data
```

```python
data = pd.concat([cleavland_data,hung_data,switzerland_data])
data
```

```python
data.columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
(data == '?').sum()
```

```python
data=data.replace('?',np.nan)
(data == '?').sum()
```

```python
data.isna().sum()
```

```python
for cols in data.columns:
    print(cols,data[cols].unique())
    print('\n')
```

```python
cols_to_mean = ['trestbps','chol','thalach','oldpeak']

# Convert specified columns to numeric, replacing non-numeric values with NaN
data[cols_to_mean] = data[cols_to_mean].apply(pd.to_numeric, errors='coerce')

# Fill NaN values with the mean of each column
data.fillna(data[cols_to_mean].mean(), inplace=True)

data[cols_to_mean] = data[cols_to_mean].abs()

# Check for missing values after conversion and filling
data.isna().sum()
```

```python
cols_to_mode = ['fbs','restecg','exang','slope','ca','thal']

data[cols_to_mode] = data[cols_to_mode].apply(pd.to_numeric, errors='coerce')

data.fillna(data[cols_to_mode].mode().iloc[0],inplace=True)

data[cols_to_mode] = data[cols_to_mode].abs()

data.isna().sum()
```

```python
for cols in data.columns:
    print(cols,data[cols].unique())
    print('\n')
```

# Data Visualization

## Histogram
### A histogram is a type of graphical representation that shows the distribution of data. It consists of a series of bars, where each bar represents a range of values, and the height of the bar indicates the frequency or count of data points falling within that range.

#### In simpler terms, histograms help you visualize how data is spread out across different value ranges. They are useful for understanding the shape, center, and spread of a dataset, as well as identifying any patterns or anomalies in the data distribution.

```python
plt.figure(figsize=(10, 7))
sns.histplot(x='age', data=data, bins=10, color='lightblue', kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

## Dot Plot
### A dot plot is a type of graph that displays data points along a number line. It is particularly useful for visualizing the distribution of a single numerical variable or comparing the distributions of several groups. Each data point is represented by a dot at its respective value on the number line.

```python
plt.figure(figsize=(10,7))
sns.stripplot(x=data['cp'],y=data['trestbps'],data=data,jitter=True,hue='cp',palette='Set1')
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Pressure')
plt.title('Chest Pain vs Resting Blood Pressure')
plt.show()
```

## Bar Plots
### A bar plot is a type of graph that presents categorical data with rectangular bars, where the length of each bar represents the proportion or count of data points within each category. Bar plots are commonly used to compare the distribution of a categorical variable across different groups or to visualize the frequency or proportion of different categories within a single variable.

```python
plt.figure(figsize=(10,7))
sns.barplot(x='cp',y='target',data=data,color='red')
plt.xlabel('Heart Disease')
plt.ylabel('Chest Pain Type')
plt.title('Heart Disease Distribution')
plt.show()
```

## Line Charts
### A line chart, also known as a line plot or line graph, is a type of graph that displays data points connected by straight lines. It is particularly useful for visualizing trends and patterns in data over time or across ordered categories. The main uses of a line chart include:

### Visualizing Trends: Line charts are effective for showing trends and patterns in data over a continuous or ordered variable, such as time, age, or another sequential variable.
### Comparing Multiple Series: Line charts can display multiple series or variables on the same plot, making it easy to compare their trends or patterns over the same time period or sequence.
### Identifying Relationships: Line charts help identify relationships between variables by showing how changes in one variable correspond to changes in another variable over time or across categories.
### Highlighting Seasonality or Cyclical Patterns: Line charts are commonly used to visualize seasonal or cyclical patterns in data, such as fluctuations in sales over different months or years.
### Communicating Changes and Variations: Line charts provide a clear visual representation of changes and variations in data, making it easy to communicate insights and findings to others.

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='age',y='target',data=data)
plt.xlabel('Age')
plt.ylabel('Target')
plt.title('Age vs Target')
plt.show()
```

## Add Histogram and Scatter plot to box plot. 
### A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers. Here's what the components of a box plot represent:

### Median (Q2): The line inside the box represents the median value of the dataset, which is the middle value when the data is sorted.
### Quartiles (Q1 and Q3): The box itself represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third ### quartile (Q3). This range covers the middle 50% of the data.
### Whiskers: The lines extending from the box (often with optional "caps" on the ends) indicate the range of the data outside the interquartile range. By default, the whiskers extend to the most extreme data points that are no more than 1.5 times the IQR from the median. Data points beyond the whiskers are considered potential outliers.
### Outliers: Individual data points that fall beyond the whiskers are plotted as individual points. These are potential outliers that may warrant further investigation.

```python
plt.figure(figsize=(10,7))

plt.subplot(2,2,1)
sns.histplot(x='age',data=data,bins=10,color='blue')
plt.xlabel('Age')
plt.ylabel('Age Distribution')

plt.subplot(2,2,2)
sns.scatterplot(x='age',y='target',data=data)
plt.xlabel('Age')
plt.ylabel('Target')

plt.subplot(2,2,3)
sns.boxplot(x=data['cp'])
plt.title('Box Plot of Cheast Pain Type')

plt.tight_layout()
plt.show()
```

## Pie Chart
### The objective of a pie chart is to visually represent the parts of a whole. It's particularly useful when you want to show the proportion of different categories or components relative to each other within a dataset.
### Showing Proportions: Pie charts help to illustrate the relative sizes of different categories or components compared to the whole.
### Highlighting Dominant Components: They make it easy to identify which categories or components contribute the most to the overall dataset.
### Simplifying Complex Data: For datasets with multiple categories or components, pie charts offer a simple and intuitive visualization method.
### Comparing Categories: Pie charts allow for easy comparison between categories, especially when the number of categories is limited.
### Communicating Key Insights: Pie charts can effectively communicate key insights or trends at a glance, making them valuable for presentations and reports.

```python
plt.figure(figsize=(8,8))
labels = ['Healthy', 'Mild', 'Moderate', 'Severe', 'Critical']
target_count = data['target'].value_counts()
plt.pie(x=target_count,data=data,labels=labels,startangle=180, autopct='%1.1f%%')
plt.title('Heart Disease Distribution')
plt.show()
```

### Whiskers: The lines extending from the box (often with optional "caps" on the ends) indicate the range of the data outside the interquartile range. ## Scatter Plot
### The objective of a scatter plot is to visualize the relationship or association between two numerical variables. It helps to identify patterns, trends, and correlations between the variables. Scatter plots are particularly useful for exploring the nature of the relationship between variables, such as whether they are positively correlated, negatively correlated, or have no correlation at all. Additionally, scatter plots can also be used to identify outliers or clusters within the data. Overall, the main objective of a scatter plot is to provide insight into the relationship between two variables and to help understand the underlying structure of the data.

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='age', y='trestbps', data=data)
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.title('Scatter Plot of Age vs Resting Blood Pressure')
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 7))
scatter_plot = sns.scatterplot(x='age', y='cp', data=data)

# Add boxplots for 'age' and 'cp' on the same plot
sns.boxplot(x='age', data=data, ax=scatter_plot.twinx(), color='skyblue', width=0.2)
sns.boxplot(y='cp', data=data, ax=scatter_plot.twiny(), color='lightgreen', width=0.2)

plt.xlabel('Age')
plt.ylabel('Chest Pain Type')
plt.title('Scatter plot with Boxplots')
plt.show()
```
