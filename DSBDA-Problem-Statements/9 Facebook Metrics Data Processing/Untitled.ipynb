# Shaping and Reshaping Data

```python
#this is done to add a column to provide a unique value to rach row to further melt and unmelt the data

data['index'] = range(1,len(data)+1)
data.head(5)
```

# Melting Data to long Format

```python
#Melting is the process of transforming a dataset from a wide format to a long format.
#In a wide format dataset, each row represents a unique observation, and each column represents a different variable or attribute.

#Unmelting is the process of transforming a dataset from a long format to a wide format.
#In a long format dataset, each row represents a single observation, and there may be multiple rows for each unique observation, each specifying a different variable or attribute.
```

```python
data_melt = pd.melt(data,id_vars=['index','Type','Category'],value_name='Value',var_name='Metrics')
data_melt
```

```python
#melting the data using melt()

data_melted = pd.melt(data, id_vars='index',  value_vars='Type')
data_melted
```

```python
data_melted = pd.melt(data, id_vars='index',  value_vars='Total Interactions')
data_melted
```
