```python
import requests as req
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/pxn-v9-adjustable-racing-steering-wheel-clutch-shifter-motion-controller/p/itmd156b160b911e?pid=ACCGTWFZBQGDHUNG&lid=LSTACCGTWFZBQGDHUNGR80BCF&marketplace=FLIPKART&q=PXN+Game+Racing&store=4rr%2Fkm5%2Fr39%2Fq8s%2Fntz&srno=s_1_2&otracker=search&otracker1=search&fm=Search&iid=90807ef7-961e-40ae-a3fe-1a95d030b626.ACCGTWFZBQGDHUNG.SEARCH&ppt=sp&ppn=sp&ssid=a754cdjnao0000001715016696065&qH=e6c6f0c69ed25122'
res = req.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text,'html.parser')
print(soup)
```

```python
product = soup.select('.VU-ZEz')
product = product[0].text
print(product)
```

```python
unedited_price = soup.select('.CxhGGd')
unedited_price = unedited_price[0].text
price = unedited_price.replace(',','')
price = price[1:]
price = float(price)
print(price)
```

```python
if(price < 10000):
    print('The Price of {} on Flipkart is in your Budget i.e {}'.format(product,price))
else:
    print('The Price of {} on Flipkart is still High than your Budget i.e 10,000'.format(product))
```
