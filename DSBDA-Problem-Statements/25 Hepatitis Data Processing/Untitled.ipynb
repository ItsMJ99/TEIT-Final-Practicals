```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./hepatitis.csv',header=None)
data
```

```python
data.columns = ['CLASS', 'AGE', 'SEX', 'STEROID', 'ANTIVIRALS', 'FATIGUE', 'MALAISE', 'ANOREXIA', 'LIVER BIG', 'LIVER FIRM', 'SPLEEN PALPABLE', 'SPIDERS', 'ASCITES', 'VARICES', 'BILIRUBIN', 'ALK PHOSPHATE', 'SGOT', 'ALBUMIN', 'PROTIME', 'HISTOLOGY']
data.head()
```

```python
data.shape
```

## Data Cleaning

```python
data.isna().sum()
```

```python
(data == '?').sum()
```

```python
for cols in data.columns:
    print(cols,data[cols].unique())
    print('\n')
```

```python
data.replace('?',np.nan,inplace=True)
```

```python
(data == '?').sum()
```

```python
data.isna().sum()
```

```python
data_to_mean = ['BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN','PROTIME']

for col in data_to_mean:
    data[col] = pd.to_numeric(data[col],errors='coerce')
    data[col] = data[col].fillna(data[col].mean())
```

```python
data.isna().sum()
```

```python
data_to_mode = ['STEROID','FATIGUE','MALAISE','ANOREXIA','LIVER BIG','LIVER FIRM','SPLEEN PALPABLE','SPIDERS','ASCITES','VARICES']

for col in data_to_mode:
    data[col] = pd.to_numeric(data[col],errors='coerce')
    data[col] = data[col].fillna(data[col].mode().loc[0])
    
```

```python
data.isna().sum()
```

```python
for col in data.columns:
    print(col,data[col].unique())
    print('\n')
```
