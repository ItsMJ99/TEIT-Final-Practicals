```python
import pandas as pd
import numpy as np
```

```python
#In Pandas, the low_memory parameter is used when reading large CSV files into a DataFrame. 
#By default, Pandas tries to determine the data types of each column automatically while reading the CSV file. 
#However, when dealing with very large files, this process can be memory-intensive.

data = pd.read_csv('../Datasets/Movie/movies_metadata.csv',low_memory=False)
data
```

# Data subsets based on Different Languages

```python
data['original_language'].unique()
```

```python
data_hi = data[data['original_language'] == 'hi']
data_hi.head(5)
```

```python
data_en = data[data['original_language'] == 'en']
data_en.head(5)
```

```python
data_mr = data[data['original_language'] == 'mr']
data_mr.head(5)
```

```python
data_ja = data[data['original_language'] == 'ja']
data_ja.head(5)
```

# Merge two subsets

```python
data_hi_mr = pd.concat([data_hi,data_mr])
data_hi_mr
```

# Sort Data using Customer Rating

```python
data_cust_rating = data.sort_values(by='vote_average',ascending=False)
data_cust_rating.head(5)
```

# Melting Data

```python
data.head(5)
```

```python
melted_data = pd.melt(data,id_vars=['id','title'],value_name='value',var_name='Metrics')
melted_data
```

```python
df_cast = pd.pivot_table(data,index=["id","original_language"],values="original_title",aggfunc="first")
df_cast_title = pd.pivot_table(data,index=["id","original_language"],values="spoken_languages",aggfunc="first")
dfr_cast_final = pd.concat([df_cast,df_cast_title],axis = 1)
```
