```python
import requests as req
from bs4 import BeautifulSoup
```

```python
url = 'https://editorial.rottentomatoes.com/guide/popular-movies/'
res = req.get(url)
html_code = res.text
print(res)
```

```python
soup = BeautifulSoup(html_code,'html.parser')
```

```python
html_prettify = soup.prettify()
print(html_prettify)
```

```python
list_title = soup.select('.article_movie_title a')
list_title = list_title[:10]

list_critic = soup.select('.critics-consensus')
list_critic = list_critic[:10]

list_synopsis = soup.select('.synopsis')
list_synopsis = list_synopsis[:10]

list_cast = soup.select('.cast')
list_cast = list_cast[:10]

list_director = soup.select('.director')
list_director = list_director[:10]

for i in range(0, 10):
    list_title[i] = list_title[i].text
    list_critic[i] = list_critic[i].text[19:]
    list_synopsis[i] = list_synopsis[i].text[10:]
    list_cast[i] = list_cast[i].text[11:]
    list_director[i] = list_director[i].text[14:]
```

```python
import pandas as pd
df = pd.DataFrame()
df['Title']=list_title
df['Critic']=list_critic
df['Description']=list_synopsis
df['Cast']=list_cast
df['Director']=list_director

df
```
