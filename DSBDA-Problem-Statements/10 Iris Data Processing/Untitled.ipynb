```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Datasets/Iris/Iris.csv')
data
```

```python
#SepalLengthCm: This column represents the length of the sepal (a part of the flower) measured in centimeters.
#SepalWidthCm: This column represents the width of the sepal measured in centimeters.
#PetalLengthCm: This column represents the length of the petal (another part of the flower) measured in centimeters.
#PetalWidthCm: This column represents the width of the petal measured in centimeters.
#Species: This column represents the species of the iris flower. In the Iris dataset, there are three species: Setosa, Versicolor, 
#and Virginica. This column serves as the target variable or class label for classification tasks, 
#where the goal is to predict the species of an iris flower based on its measurements.
```

```python
data.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
data.head(5)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.describe()
```

# Species Subsets

```python
data['Species'].unique()
```

```python
data_iris_setosa = data[data['Species']=='Iris-setosa']
data_iris_setosa.head(5)
```

```python
data_iris_versicolor = data[data['Species']=='Iris-versicolor']
data_iris_versicolor.head(5)
```

```python
data_iris_virginica = data[data['Species']=='Iris-virginica']
data_iris_virginica.head(5)
```

# Merge two subsets

```python
data_iris_versicolor_virginica = pd.concat([data_iris_versicolor,data_iris_virginica])
data_iris_versicolor_virginica
```

# Sort Data on Petal Length

```python
data_petal_length = data.sort_values(by='PetalLengthCm',ascending=False)
data_petal_length
```

# Data Transpose

```python
data_transpose = data.transpose()
data_transpose
```

# Data Melting

```python
data.head(5)
```

```python
melted_data = pd.melt(data,id_vars=['Species','SepalLengthCm','PetalLengthCm'],var_name='Sepal Width',value_name='in Cm')
melted_data
```

# Cast data to wide format

```python
casted_data = pd.pivot_table(data,index='Species')
casted_data
```
