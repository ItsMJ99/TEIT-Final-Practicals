```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./adult.csv',header=None)
data
```

```python
data.columns = ['age','workclass','fntwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
data
```

```python
(data == ' ?').sum()
```

```python
for cols in data.columns:
    print(cols,data[cols].unique())
    print('\n')
```

```python
cols_whitespaces = ['workclass','education','marital-status','occupation','relationship','race','sex','native-country']

for cols in cols_whitespaces:
    data[cols] = data[cols].str.strip()
```

```python
for cols in data.columns:
    print(cols,data[cols].unique())
    print('\n')
```

```python
cols_to_mode = ['workclass','occupation','native-country']
```

```python
data.replace('?',np.nan,inplace=True)
```

```python
for cols in cols_to_mode:
    data[cols] = data[cols].fillna(data[cols].mode()[0])

(data == '?').sum()
```

# Data Visualization

## Histogram
### Objective: To visualize the distribution of a single continuous variable, such as age, education-num, capital-gain, or capital-loss. Histograms help us understand the underlying frequency distribution of the data and identify any patterns or outliers

```python
plt.figure(figsize=(10,7))
sns.histplot(x='age',data=data,bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

## Dot Plot
### Objective: Dot plots can be used to compare the values of a continuous variable across different categories. For example, you could create a dot plot showing the distribution of ages within different education levels or work classes. Dot plots help in visualizing the spread and central tendency of the data within each category.

```python
plt.figure(figsize=(10,7))
sns.stripplot(x='workclass',y='age',data=data,hue='workclass',palette='Set1')
plt.xlabel('Workclass')
plt.ylabel('Age')
plt.title('Workclass vs Age')
plt.show()
```

## Bar Plot
### Objective: Bar plots are useful for comparing the frequency or distribution of a categorical variable across different categories. You could create bar plots to show the number of individuals in each income category for different native countries or races. Bar plots help in visualizing the distribution of categorical data and identifying any patterns or discrepancies.

```python
#calculates the mean age for each combination of 'native-country' and 'income', and country_age = country_age.reset_index() 
#resets the index of the resulting DataFrame to make it more manageable.

country_age = data[['native-country','income','age']].groupby(['native-country','income']).mean()
country_age = country_age.reset_index()
```

```python
plt.figure(figsize=(20,20))
sns.barplot(x='age',y='native-country',data=country_age,hue='native-country',palette='Set1')
plt.xlabel('Age')
plt.ylabel('Country')
plt.title('Mean Age with Income by Country')
plt.show()
```

## Line Chart
### Objective: Line charts are typically used to visualize trends over time. However, in this dataset, since we don't have a time-related variable, we can use line charts to show the relationship between two continuous variables, such as age and hours-per-week. Line charts help in understanding the relationship or correlation between two continuous variables.

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='age',y='hours-per-week',data=data)
plt.xlabel('Age')
plt.ylabel('Hours Per Week')
plt.title('Age vs Hours Per Week')
plt.show()
```

# Add Histogram and Scatter plot to box plot

```python
plt.figure(figsize=(10,7))

#histogram
plt.subplot(2,2,1)
sns.histplot(x='age',data=data,bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

#scatterplot
plt.subplot(2,2,2)
sns.scatterplot(x='age', y='education-num',data=data)
plt.xlabel('Age')
plt.ylabel('Education Number')
plt.title('Age vs Education Number')

#boxplot
plt.subplot(2,2,3)
plt.boxplot(x='hours-per-week',data=data)
plt.title('Hours Per Week Working')

plt.tight_layout()
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(15,7))
sns.boxplot(x='education', y='age', data=data)
plt.xlabel('Education')
plt.ylabel('Age')
plt.title('Age Distribution by Education Level')
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='age', y='education-num', data=data)
plt.xlabel('Age')
plt.ylabel('Education Number')
plt.title('Scatter Plot of Age vs Education Number')
plt.show()
```

## Add boxplots to a scatterplot

```python
plt.figure(figsize=(20, 17))

scatter = sns.scatterplot(x='age', y='education-num', data=data)
# Segment the scatterplot based on a categorical variable (e.g., 'income') and plot boxplots for each segment
sns.boxplot(x='age', y='education-num', data=data,width=0.2)

plt.xlabel('Age')
plt.ylabel('Education Number')
plt.title('Scatter Plot with Boxplots')
plt.show()
```
